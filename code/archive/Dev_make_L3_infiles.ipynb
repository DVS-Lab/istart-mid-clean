```python
print("subs total: %s"%(len(subs)))

cope_list=natsorted([re.search('(.*).feat',cope).group(1) for cope
                     in os.listdir(
    '/data/projects/istart-mid/derivatives/fsl/sub-1001/L2_task-mid_model-1_type-act.gfeat/')
                    if cope.startswith('cope')])
print("number of copes: %s"%(len(cope_list)))
for cope in cope_list:
    print(cope)
    oneR_subs=[]
    twoR_subs=[]
    for sub in subs:
        
        datanameL2='L2_task-mid_model-4_type-act.gfeat/%s.feat/stats/cope1.nii.gz'%(cope)
        datanameL1='L1_task-mid_model-4_type-act_run-1.feat/stats/%s.nii.gz'%(cope)
        if os.path.exists(fsldir+'/sub-%s/%s'%(sub,datanameL2)):
            temp_file='L2_task-mid_model-1_type-REPLACEME.gfeat/copeCOPENUM.feat/stats/cope1.nii.gz'
            twoR_subs.append(fsldir+'/sub-%s/%s'%(sub,temp_file))
        
        elif os.path.exists(fsldir+'/sub-%s/%s'%(sub,datanameL1)):
            temp_file='/L1_task-mid_model-1_type-REPLACEME_run-1.feat/stats/copeCOPENUM.nii.gz'
            oneR_subs.append(fsldir+'/sub-%s/%s'%(sub,temp_file))
           
        else:
            print("%s doesn't have recognizale output"%(sub))
    fnames=natsorted(oneR_subs + twoR_subs)
    df=pd.DataFrame(data=fnames,columns=['fnames'])
    df.to_csv('/data/projects/istart-mid/code/L3_data_files/L3_model-4_type-act_Inputdata_%s.tsv'%(cope),
              sep='\t',
              index=False)
display(df)

```
